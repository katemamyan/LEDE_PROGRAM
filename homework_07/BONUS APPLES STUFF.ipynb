```python
import requests
from bs4 import BeautifulSoup

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36" 
}
response = requests.get('https://www.applesfromny.com/varieties/', headers=headers)
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
# doc.find_all('div', class_='apple-variety')
# doc.find_all('div', { 'class': 'apple-variety' })
varieties = doc.select(".apple-variety")

rows = []
for variety in varieties:
    # print("-----")
    row = {}
    # print(variety)
    row['title'] = variety.select_one("h2").text
    row['sweetness'] = variety.select_one(".scale-icon").get('class')[-1]
    row['desc'] = variety.select_one("p").text
    row['url'] = variety.select_one("a").get('href')
    row['img_url'] = variety.select_one("img").get('src')
    rows.append(row)
len(rows)
```

```python
import pandas as pd

apples = pd.DataFrame(rows)
apples
```
